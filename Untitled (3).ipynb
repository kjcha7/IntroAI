```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1단계: 데이터 준비

```python
train = pd.read_csv("titanic_train.csv", index_col="PassengerId")
print(train.shape)
train.head()
```

```python
test = pd.read_csv("titanic_test.csv", index_col="PassengerId")
print(test.shape)
test.head()
```

## 2단계 : EDA

```python
train["Embarked"].value_counts()
```

```python
%matplotlib inline
```

```python
sns.countplot(data=train, x="Embarked", hue="Survived")
```

```python
train.corr()
```

```python
embarked = pd.get_dummies(train["Embarked"], prefix="Embarked")
embarked
```

```python
test_embarked = pd.get_dummies(test["Embarked"], prefix="Embarked")
test_embarked
```

```python
test = pd.concat([test, test_embarked], axis=1)
test.head()
```

```python
train = pd.concat([train, embarked], axis=1)
train.head()
```

```python
train.corr()
```

```python
sns.countplot(data=train, x="Sex", hue="Survived")
```

```python
sns.distplot(train["Fare"])
```

```python
ignore_high_fare = train[train["Fare"] < 100]
```

```python
survived = ignore_high_fare[ignore_high_fare.Survived == 1]
perish = ignore_high_fare[ignore_high_fare.Survived == 0]
print(survived.shape)
print(perish.shape)
```

```python
sns.distplot(survived["Fare"], hist=False, label="Survivor")
sns.distplot(perish["Fare"], hist=False, label="Perish")
```

```python
train.loc[train.Sex == "male", "Sex_encode"] = 0
train.loc[train.Sex == "female", "Sex_encode"] = 1
train.head()
```

```python
test.loc[test.Sex == "male", "Sex_encode"] = 0
test.loc[test.Sex == "female", "Sex_encode"] = 1
test.head()
```

```python
sns.countplot(data=train, x="Parch", hue="Survived")
```

```python
test.Fare.isnull().sum()
```

```python
mean_fare = train["Fare"].mean()
mean_fare
```

```python
test.loc[test.Fare.isnull(), "Fare"] = mean_fare
```

```python
test[test.Fare.isnull()]
```

## 예측모형

```python
train.columns
```

```python
x_values = ["Pclass", "Sex_encode", "Fare", "Embarked_C",
           "Embarked_Q", "Embarked_S"]
```

```python
y_train = train["Survived"]
```

```python
X_train = train[x_values]
X_test = test[x_values]
```

- 모델선택

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
model_A = DecisionTreeClassifier(max_depth = 5)
```

```python
model_B = RandomForestClassifier(n_estimators = 3000, 
                                 max_depth=7,
                                random_state=31)
```

- 학습

```python
model_A.fit(X_train, y_train)
```

```python
model_B.fit(X_train, y_train)
```

- 예측

```python
prediction_A = model_A.predict(X_test)
prediction_A
```

```python
prediction_B = model_B.predict(X_test)
prediction_B
```

- 제출

```python
submission = pd.read_csv("gender_submission.csv", index_col="PassengerId")
submission["Survived"] = prediction_B
```

```python
submission.to_csv("model_B.csv")
```
